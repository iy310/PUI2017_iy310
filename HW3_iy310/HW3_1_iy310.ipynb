```python
mymean = 100
```

```python
normal_li = []
poisson_li = []
binomial_li = []
chisq_li = []
rayleigh_li = []
```

```python
normalX = []
normalY = []
poissonX = []
poissonY = []
binomialX = []
binomialY = []
chisqX = []
chisqY = []
rayleighX = []
rayleighY = []

for i in range(len(normal_li)):
    normalY.append(np.mean(normal_li[i]))
    normalX.append(len(normal_li[i]))
for i in range(len(poisson_li)):
    poissonY.append(np.mean(normal_li[i]))
    poissonX.append(len(poisson_li[i]))
for i in range(len(binomial_li)):
    binomialY.append(np.mean(binomial_li[i]))
    binomialX.append(len(binomial_li[i]))
for i in range(len(chisq_li)):
    chisqY.append(np.mean(chisq_li[i]))
    chisqX.append(len(chisq_li[i]))
for i in range(len(rayleigh_li)):
    rayleighY.append(np.mean(rayleigh_li[i]))
    rayleighX.append(len(rayleigh_li[i]))
```
