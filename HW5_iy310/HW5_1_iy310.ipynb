```python
plt.title("Distribution")
plt.hist(samples[0])
plt.xlabel("X Values")
plt.ylabel("Frequency")
plt.show()
```

```python
zs = []
for i in range(len(samples)):
    dev = np.std(samples[i])
    mean = np.mean(samples[i])
    zs.append((mean - 10) * np.sqrt(n_size) / dev )
```

```python
param = scipy.stats.norm.fit(zs)
x = np.linspace(-5,5,100)
pdf_fitted = scipy.stats.norm.pdf(x,loc=param[0],scale=param[1])
```

```python
plt.title('Z-Test')
plt.plot(x, pdf_fitted,'r-') 
plt.hist(zs, normed=1)
plt.xlabel("Standard Dev")
plt.ylabel("Frequency")
plt.show()
```

```python
ks = scipy.stats.kstest(zs, 'norm')
print(ks)
```


The P-value shows that the z-test distribution is same as that of the normal distribution, thereby unable to reject the null-hypothesis.
